```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(0,np.pi+0.01,np.pi/30)
y = np.cos(t)

S = np.sum(t * y)

print(f"The sum is: {S}")
```

```python
import random

# Define the parametric equations for x(θ) and y(θ)
def x_theta(theta, R, delta_r, f, p):
    return R * (1 + delta_r * np.sin(f * theta + p)) * np.cos(theta)

def y_theta(theta, R, delta_r, f, p):
    return R * (1 + delta_r * np.sin(f * theta + p)) * np.sin(theta)

# First figure parameters
R = 1.2
delta_r = 0.1
f = 15
p = 0
theta = np.linspace(0, 2 * np.pi, 1000)

# Compute x and y for the first curve
x1 = x_theta(theta, R, delta_r, f, p)
y1 = y_theta(theta, R, delta_r, f, p)

# Plot the first figure (wavy circle)
plt.figure(figsize=(6, 6))
plt.plot(x1, y1)
plt.title("Wavy Circle")
plt.axis('equal')
plt.grid(True)

# Second figure with for loop
plt.figure(figsize=(6, 6))

for i in range(1, 11):
    R_i = i
    delta_r_i = 0.05
    f_i = 2 + i
    p_i = random.uniform(0, 2 * np.pi)  # Random p value between 0 and 2π
    
    # Compute x and y for the ith curve
    x_i = x_theta(theta, R_i, delta_r_i, f_i, p_i)
    y_i = y_theta(theta, R_i, delta_r_i, f_i, p_i)
    
    # Plot each curve
    plt.plot(x_i, y_i)

plt.title("Multiple Wavy Circles with Random Phase")
plt.axis('equal')
plt.grid(True)

# Show both figures
plt.show()
```
